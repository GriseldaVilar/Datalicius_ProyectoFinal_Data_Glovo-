```python
#Mirar si en la columna Price podemos completar info con la media de la columna "Neighbourhood Cleansed", teniendo en cuenta el barrio 
df_madrid_neig_clean=(
    df_madrid.groupby('Neighbourhood Cleansed')
    .agg({'Price':'mean', 'ID':'count'})
    .sort_values(by='Price', ascending=False)
)
df_madrid_neig_clean
```


#Comprobamos que si utilizamos la columna "Neighbourhood Group Cleansed", los precios de la media se nos disparan menos que con la columna "Neighbourhood Cleans", ya que tiene mas datos 

```python
#Crear una nueva columnna "Price_new", donde los valores nulos los sustituimos por la media de la columna "Neighbourhood Group Cleansed"
df_madrid2= (
    df_madrid
    .assign(
        Price_new = lambda df_:df_['Price'].fillna(df_.groupby('Neighbourhood Group Cleansed')['Price'].transform('mean'))
    )
)
df_madrid2['Price_new']

```

```python
#Completamos los valores nulos del Weekly Price, multiplicando la columna "Price_new" por 7(dias de la semana)
df_madrid2= (
    df_madrid2
    .assign(
        Price_new_weekly = lambda df_:df_['Weekly Price'].fillna(df_['Price_new']*7)
    )
)
df_madrid2['Price_new_weekly']
```

```python
#Completamos los valores nules del Monthly Price, multiplicando la columna "Price_new" por 30(dias del mes)
df_madrid2= (
    df_madrid2
    .assign(
        Price_new_monthly = lambda df_:df_['Monthly Price'].fillna(df_['Price_new']*30)
    )
)
df_madrid2['Price_new_monthly']
```

Completar los valores NaN de las columnas Bathrooms, Beds, Bedrooms con la mediana

```python
#Column BEDS
df_madrid_beds=(
    df_madrid2.groupby('Property Type')
    .agg({'Beds':'mean', 'ID':'count'})
    .sort_values(by='Beds', ascending=False)
    .round(0)
)
df_madrid_beds
```

```python
#Creamos una nueva columnna "Beds_new", donde los valores nulos los sustituimos por la media de la columna "Property Type"
df_madrid2= (
    df_madrid2
    .assign(
        Beds_new= lambda df_:df_['Beds'].fillna(df_.groupby('Property Type')['Beds'].transform('mean'))
    )
)
df_madrid2['Beds_new']
```

```python
#Columns Bedrooms
df_madrid_bedrooms=(
    df_madrid2.groupby('Property Type')
    .agg({'Bedrooms':'mean', 'ID':'count'})
    .sort_values(by='Bedrooms', ascending=False)
    .round(0)
)
df_madrid_bedrooms
```

```python
#Crear una nueva columnna "Bedrooms_new", donde los valores nulos los sustituimos por la media de la columna "Property Type"
df_madrid2= (
    df_madrid2
    .assign(
        Bedrooms_new= lambda df_:df_['Bedrooms'].fillna(df_.groupby('Property Type')['Bedrooms'].transform('mean'))
    )
)
df_madrid2['Bedrooms_new']
```

```python
#Bathrooms
df_madrid_bathrooms=(
    df_madrid2.groupby('Property Type')
    .agg({'Bathrooms':'mean', 'ID':'count'})
    .sort_values(by='Bathrooms', ascending=False)
    .round(0)
)
df_madrid_bathrooms
```

```python
#Crear una nueva columnna "Bathrooms_new", donde los valores nulos los sustituimos por la media de la columna "Property Type"
df_madrid2= (
    df_madrid2
    .assign(
        Bathrooms_new= lambda df_:df_['Bathrooms'].fillna(df_.groupby('Property Type')['Bathrooms'].transform('mean'))
    )
)
df_madrid2['Bathrooms_new']
```
